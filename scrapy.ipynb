## Webscrapping Newegg.com Most Popular Graphics Cards


```python

import scrapy


class QuotesSpider(scrapy.Spider):
    name = "quotes"

    def start_requests(self):
        urls = [
            'https://www.newegg.com/Desktop-Graphics-Cards/SubCategory/ID-48?Tid=7709&PageSize=96&Order=3'
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        pg_title0 = response.css('title::text').extract()
        
        for item in response.css('div.item-container'):
            link = item.xpath('./a/@href').get()
            title = item.xpath('./a/img/@title').get()
            brand = item.xpath('.//div[@class="item-branding"]/a/img/@title').get()
            rating = item.xpath('.//div[@class="item-branding"]/a/i/@class').gett()
            nratings=item.xpath('.//div[@class="item-branding"]/a/span/text()').get()
            price1 = item.css('li.price-current>strong::text').get()
            price2 = item.css('li.price-current>sup::text').get()
            print(brand,  (str(price1)+str(price2)).replace(",",""), title)
            yield {
                'brand' : brand,
                'rating' : rating,
                "price" : (str(price1)+str(price2)).replace(",",""),
                'title' : title,
                'link' : link
                }
        
        """filename = 'quotes-%s.html' % page
        with open(filename, 'wb') as f:
            f.write(response.body)
        self.log('Saved file %s' % filename)
        """

```
